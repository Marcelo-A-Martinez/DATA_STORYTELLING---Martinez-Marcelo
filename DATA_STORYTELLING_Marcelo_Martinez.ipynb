El siguiente conjunto de datos corresponde a los registros obtenidos del estado de Washington, EE.UU entre abril del 2020 y marzo del 2021.

```python
Casos= pd.read_excel('C:\\Users\\USUARIO\\DATA SCIENCE - PROYECTO\\CASOS\\Casos.xlsx' , header=0 )
Casos.head()
```

Durante el periodo analizado la cantidad tanto de casos positivos como negativos fue incrementándose notablemente. 
Una particularidad a resaltar, es que la cantidad de casos encuadrados como "Negativos" comenzó a incrementarse más que proporcionalmente respecto a los "Positivos".
Esto se debe a que justamente, la cantidad de casos "Negativos" fue incrementándose día a día, salvo días específicos donde esta decrece a niveles inferiores a los casos "Positivos". De esta manera se puede indicar que el día con la mayor variación negativa de casos "Negativos" se da el 2 de diciembre del 2020 y el máximo se produce el día 5 de enero del 2021.
En cuanto a los casos "Positivos", la mayor variación se da el día 8 de enero del 2021 y la menor variación el día 26 de mayo del 2020.
Los picos mencionados pueden deberse a las festividades que se celebran en esas fechas, ya que esto genera un movimiento mayor de personas que buscan reunirse con sus familiares para pasar juntos durante las fiestas. 

```python
neg = sns.lineplot(data=Casos, x='Fecha', y='Negativos', label='Negativos')
pos = sns.lineplot(data=Casos, x='Fecha', y='Positivos' , label='Positivos')
plt.title("Casos Positivos y Negativos")
plt.xlabel("Meses")
plt.ylabel("Cantidades")
plt.legend()
```

```python
inc_neg = sns.lineplot(data=Casos, x='Fecha', y='Aumento_de_Negativos', label='Incremento Negativos')
inc_pos = sns.lineplot(data=Casos, x='Fecha', y='Aumento_Positivos' , label='Incremento Positivos')
plt.title("Incremento de casos Positivos y Negativos")
plt.xlabel("Meses")
plt.ylabel("Cantidades")
plt.legend()
```

### DECESOS

En cuanto a la cantidad de muertes diarias que se dieron en el periodo analizado, se puede observar que en los primeros meses de la pandemia hubo un número elevado de decesos. 
Entre junio y noviembre del 2020, los niveles de mortalidad de pacientes con COVID se mantuvieron relativamente estables.
Esto comienza a cambiar a partir de diciembre, donde se puede observar que la cantidad de personas que fallecen a causa de este virus comienza a crecer llegando a un máximo de 5.427 el día 12 de febrero del 2021.
A partir de allí, se pude visualizar un decremento en la cantidad de fallecimientos.

```python
decesos = sns.lineplot(data=Casos, x='Fecha', y='Incremento_de_Muertes')
plt.title("Evolución de Decesos de pacientes con COVID-19")
plt.xlabel("Meses")
plt.ylabel("Cantidades")
```

### HOSPITALIZADOS

```python
Hosp = sns.lineplot(data=Casos, x='Fecha', y='Hospitalizado_Actualmente')
plt.title("Hospitalizados por día")
plt.xlabel("Días")
plt.ylabel("Cantidades")
```
